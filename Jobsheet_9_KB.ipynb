```python
# Import library
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline

# Load dataset
df = pd.read_csv('diabetes.csv')
X = df.drop('Outcome', axis=1)
y = df['Outcome']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Pipeline dengan StandardScaler dan SVM
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('svm', SVC())
])

# Parameter grid untuk GridSearch
param_grid = {
    'svm__C': [0.1, 1, 10, 100],              # Parameter regularisasi
    'svm__kernel': ['linear', 'rbf', 'poly'],  # Jenis kernel
    'svm__gamma': ['scale', 'auto', 0.001, 0.01, 0.1]  # Koefisien kernel (untuk rbf/poly)
}

# Grid Search dengan 5-fold cross-validation
grid_search = GridSearchCV(
    pipeline,
    param_grid,
    cv=5,
    scoring='accuracy',
    verbose=1,
    n_jobs=-1  # Gunakan semua core CPU
)

# Fit model
grid_search.fit(X_train, y_train)

# Hasil terbaik
print("Parameter Terbaik:", grid_search.best_params_)
print("Akurasi Training:", grid_search.best_score_)
print("Akurasi Test:", grid_search.score(X_test, y_test))
```

```python
# Import library
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error

# Load dataset
data = pd.read_csv('Salary_Data.csv')
X = data[['YearsExperience']]
y = data['Salary']

# Parameter grid untuk SVR
param_grid = {
    'C': [100, 500, 1000],          # Parameter regularisasi
    'kernel': ['rbf', 'linear'],
    'gamma': ['scale', 'auto', 0.01, 0.1],
    'epsilon': [0.1, 0.5, 1]        # Parameter toleransi error
}

# Grid Search
grid_search = GridSearchCV(
    SVR(),
    param_grid,
    cv=5,
    scoring='neg_mean_squared_error',  # Metric untuk regression
    verbose=1,
    n_jobs=-1
)

# Fit model
grid_search.fit(X, y)

# Hasil terbaik
print("Parameter Terbaik:", grid_search.best_params_)
print("MSE:", -grid_search.best_score_)  # MSE (semakin kecil semakin baik)

# Visualisasi hasil terbaik
best_model = grid_search.best_estimator_
plt.scatter(X, y, color='blue')
plt.plot(X, best_model.predict(X), color='red')
plt.title('SVR dengan Parameter Optimal')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# Import library yang diperlukan
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import MinMaxScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report

# 1. Load Dataset
try:
    dataset = pd.read_csv('iklan_sosmed.csv', delimiter=';')
except FileNotFoundError:
    print("Error: File tidak ditemukan. Pastikan 'iklan_sosmed.csv' ada di direktori yang benar.")
    exit()

# 2. Preprocessing Data
X = dataset.iloc[:, 2:-1].values  # Ambil kolom umur dan gaji
y = dataset.iloc[:, -1].values    # Kolom target (Transaksi)

# Split data training (80%) dan testing (20%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=99)

# Normalisasi data
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# 3. Baseline Model (Tanpa Tuning)
print("\n=== Baseline Model ===")
baseline_model = GaussianNB()
baseline_model.fit(X_train, y_train)
y_pred = baseline_model.predict(X_test)
baseline_acc = accuracy_score(y_test, y_pred)
print(f"Akurasi Baseline: {baseline_acc*100:.2f}%")
print(classification_report(y_test, y_pred))

# 4. Hyperparameter Tuning dengan GridSearchCV
print("\n=== Hyperparameter Tuning ===")

# Definisikan parameter grid
param_grid = {
    'var_smoothing': np.logspace(-12, 0, num=100)  # Rentang nilai untuk var_smoothing
}

# Buat model dengan GridSearchCV
tuned_model = GridSearchCV(
    estimator=GaussianNB(),
    param_grid=param_grid,
    cv=5,               # 5-fold cross validation
    scoring='accuracy',  # Metric evaluasi
    n_jobs=-1,          # Gunakan semua core CPU
    verbose=1
)

# Latih model dengan parameter terbaik
tuned_model.fit(X_train, y_train)

# 5. Evaluasi Model Hasil Tuning
print("\n=== Hasil Tuning ===")
print(f"Parameter Terbaik: {tuned_model.best_params_}")
print(f"Akurasi Cross-Validation Terbaik: {tuned_model.best_score_*100:.2f}%")

y_pred_tuned = tuned_model.predict(X_test)
tuned_acc = accuracy_score(y_test, y_pred_tuned)
print(f"\nAkurasi pada Test Set: {tuned_acc*100:.2f}%")
print(classification_report(y_test, y_pred_tuned))

# 6. Perbandingan Hasil
improvement = (tuned_acc - baseline_acc) * 100
print(f"\n=== Perbandingan ===")
print(f"Peningkatan Akurasi: {improvement:.2f}%")
print(f"Baseline: {baseline_acc*100:.2f}% → Tuned: {tuned_acc*100:.2f}%")

# 7. Visualisasi Decision Boundary (Opsional)
def plot_decision_boundary(X, y, model, title):
    plt.figure(figsize=(10, 6))
    X_set, y_set = X, y

    X1, X2 = np.meshgrid(
        np.arange(start=X_set[:, 0].min()-0.1, stop=X_set[:, 0].max()+0.1, step=0.01),
        np.arange(start=X_set[:, 1].min()-0.1, stop=X_set[:, 1].max()+0.1, step=0.01)
    )

    plt.contourf(X1, X2, model.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
                alpha=0.5, cmap=ListedColormap(('red', 'green')))

    for i, j in enumerate(np.unique(y_set)):
        plt.scatter(X_set[y_set==j, 0], X_set[y_set==j, 1],
                   c=ListedColormap(('red', 'green'))(i), label=j, edgecolor='black')

    plt.title(title)
    plt.xlabel('Umur (Normalized)')
    plt.ylabel('Gaji (Normalized)')
    plt.legend()
    plt.show()

# Plot hasil tuning
plot_decision_boundary(X_train, y_train, tuned_model.best_estimator_, "Decision Boundary (Model Hasil Tuning)")
```
